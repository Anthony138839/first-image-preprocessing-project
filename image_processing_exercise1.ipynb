**Converting RGB image to Grayscale using OpenCV**

```python
# Loading the image
img = cv2.imread(r"C:\Users\USER\Documents\EMEKA'S CODING STUFF\BACKEND\PYTHON IN GENERAL\Data Science\Introduction To Machine Learning\Codes\Project 22\dice.jpg")
```

```python
# Converting the image to Grayscale
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Plotting the image
plt.imshow(grayscale_img, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Saving the grayscale image
cv2.imwrite("dice grayscale image.jpg", grayscale_img)
```
